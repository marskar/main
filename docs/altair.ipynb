# Plotting with Altair

## Introduction
Altair is a Python visualization library built on the [Vega visualization grammar](https://vega.github.io/vega/).
Altair strives to be a declarative, rather than imperative, plotting system.
This allows Altair to make compelling plots without being provided detailed step-by-step instructions.

## Example
The plot below is from the [Altair Example Gallery](https://altair-viz.github.io/gallery/index.html).

```python
import altair as alt
from vega_datasets import data

source = data.cars()

alt.Chart(source).mark_circle(size=60).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    tooltip=['Name', 'Origin', 'Horsepower', 'Miles_per_Gallon']
).interactive()
```

## Conclusion
As you can see from the code and corresponding plot above, Altair is succinct and powerful!
